# Insight 1: Customer Demographics

```python
Customer_region_counts = df['Region'].value_counts()
Customer_region_counts = df['Region'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=Customer_region_counts.index, y=Customer_region_counts.values)
plt.title('Customer Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.show()
```

# Insight 2: Product Pricing Strategy

```python
plt.figure(figsize=(10, 6))
sns.histplot(df1['Price'], bins=30, kde=True)
plt.title('Product Price Distribution')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()
```

# Insight 3: Seasonal Sales Trends

```python
df2['TransactionDate'] = pd.to_datetime(df2['TransactionDate'])
monthly_transactions = df2.resample('M', on='TransactionDate').size()
plt.figure(figsize=(12, 6))
monthly_transactions.plot()
plt.title('Total Transactions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.show()
```

# Insight 4: Correlation Between Quantity and Revenue

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df2, x='Quantity', y='TotalValue')
plt.title('Quantity vs Total Value')
plt.xlabel('Quantity Purchased')
plt.ylabel('Total Transaction Value (USD)')
plt.show()
```

# Insight 5: Product Category Performance

```python
category_performance = df2.merge(df1, on='ProductID').groupby('Category')['TotalValue'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=category_performance.index, y=category_performance.values)
plt.title('Total Revenue by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue (USD)')
plt.show()
```
